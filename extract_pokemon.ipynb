```python
import requests
import pandas as pd
import numpy as np
```

```python
df_pokemon = pd.DataFrame(columns = ['id', 'name', 'base_xp', 'height', 'weight', 'nabilities', 'abilities', 'meanstats', 'stats', 'ntypes', 'types', 'islegendary'])


for i in range(1, 899):
    
    #Requisições na API
    
    api = f'https://pokeapi.co/api/v2/pokemon/{i}'
    res = requests.get(api)
    poke = res.json()

    api2 = f'https://pokeapi.co/api/v2/pokemon-species/{i}'
    res2 = requests.get(api2)
    poke2 = res2.json()
    
    # Infos básicas

    id = poke['id']
    name = poke['name']
    base_xp = poke['base_experience']
    heightcm = poke['height'] * 10
    weightkg = poke['weight'] / 10


    # Habilidades

    abilities = poke['abilities']
    abilities = [ability['ability']['name'] for ability in abilities]
    nabilities = len(abilities)

    # Status Base
    x = poke['stats']
    stats = {}
    for stat in x:
        stats[stat['stat']['name']] = stat['base_stat']
    meanstats = int(sum(stats.values())/len(stats))

    # Tipos
    types = poke['types']
    types = [_['type']['name'] for _ in types]
    ntypes = len(types)

    # É Lendário
    if poke2['is_legendary'] or poke2['is_mythical']: islegendary = 1
    else: islegendary = 0
        
    pokemon = {'id':id,
           'name':name,
           'base_xp':base_xp,
           'height':heightcm,
           'weight':weightkg, 
           'nabilities':nabilities,
           'abilities':abilities,
           'meanstats':meanstats,
           'stats':stats,
           'ntypes':ntypes,
           'types':types,
           'islegendary':islegendary}
    
    df_pokemon = df_pokemon.append(pokemon, ignore_index=True)

df_pokemon = df_pokemon.set_index('id')
```
